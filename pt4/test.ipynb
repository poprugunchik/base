```python
my_list = [9,12,1,5,11]
def change (my_list):
    for i in range(1, len(my_list)):
        key = my_list[i]
        j = i - 1
        while j >= 0 and my_list[j] > key:
            my_list[j + 1] = my_list[j]
            j -= 1
        my_list[j + 1] = key
    return my_list
print(change(my_list))
```

```python
# Для этих ожидаем успешный логин
USER_LOGIN = 'login'
USER_PASSWORD = 'password'

# Мамкиного хацкера отфутболиваем (логин может совпадать, но пароль другой)
HACKER_LOGIN = 'login'
HACKER_PASSWORD = 'hacker_password'

# Простая "база" пользователей в памяти
_users = {}


def register(username: str, password: str) -> str:
    """Регистрирует пользователя и возвращает username.
    Если пользователь уже существует — перезаписывает пароль (чтобы тесты были детерминированы).
    """
    username = username.strip()
    password = password.strip()
    if not username:
        raise ValueError("username не может быть пустым")
    _users[username] = password
    return username


def login(username: str, password: str) -> str | None:
    """Проверяет логин/пароль. Возвращает username при успешном входе, иначе None."""
    username = username.strip()
    password = password.strip()
    stored = _users.get(username)
    if stored is None:
        return None
    return username if stored == password else None


# Первый кейс - рега и логин из переменных
def test_with_vars() -> None:
    new_user = register(USER_LOGIN, USER_PASSWORD)  # зарегали
    login_user = login(USER_LOGIN, USER_PASSWORD)  # т.к. юзер зареган — успешно входит
    hacker_login = login(HACKER_LOGIN, HACKER_PASSWORD)  # пытался зайти с неправильным паролем

    # Проверяем
    if new_user == login_user:
        print("Логин работает")
    else:
        print("Что-то не так: зарегистрированный пользователь не может войти")

    if hacker_login is None:
        print("Хакер не пройдет")
    else:
        print("Проблема: хакер вошёл!")


# Второй кейс - рега и логин из input()
def test_register_with_input() -> str:
    print("=== Регистрация через input ===")
    username_input = input("Введите логин для регистрации: ").strip()
    password_input = input("Введите пароль для регистрации: ").strip()
    new_user = register(username_input, password_input)
    print(f"Зарегали нового юзера: {new_user}")
    return new_user


def test_login_with_input(expected_username: str | None = None) -> None:
    print("=== Попытки логина через input (введите данные) ===")
    while True:
        username_try = input("Логин: ").strip()
        password_try = input("Пароль: ").strip()
        user = login(username_try, password_try)
        if user:
            # если передали expected_username — показываем его, иначе показываем вошедшего юзера
            if expected_username:
                print(f"Логиним юзера {expected_username}")
            else:
                print(f"Логиним юзера {user}")
            break
        else:
            print("Ты не пройдешь! Попробуй ещё раз.")


if __name__ == '__main__':
    # автоматический тест с переменных
    test_with_vars()

    # интерактивная регистрация + логин
    registered = test_register_with_input()
    test_login_with_input(expected_username=registered)
```

```python
def register(username, password, filename="users.txt"):
    try:
        with open(filename, "r", encoding="utf-8") as f:
            names = {line.split(":",1)[0].strip() for line in f}
    except FileNotFoundError:
        names = set()
    if username in names:
        return False
    with open(filename, "a", encoding="utf-8") as f:
        f.write(f"{username}:{password}\n")
    return True


def login_function(username, password, filename="users.txt"):
    try:
        with open(filename, "r", encoding="utf-8") as f:
            for line in f:
                saved_username, saved_password = line.strip().split(":", 1)
                if saved_username == username and saved_password == password:
                    return True
    except FileNotFoundError:
        return False
    return False
```
